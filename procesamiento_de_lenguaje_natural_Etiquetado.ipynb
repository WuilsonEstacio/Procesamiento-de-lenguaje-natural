##Etiquetado en Español
Para el ingles, NLTK tiene tokenizador y etiquetador pre-entrenados por defecto. En cambio, para otros idiomas es preciso entrenarlo previamente.

usamos el corpus cess_esp https://mailman.uib.no/public/corpora/2007-October/005448.html

el cual usa una convención de etiquetas gramaticales dada por el grupo EAGLES https://www.cs.upc.edu/~nlp/tools/parole-sp.html

```python
import nltk
nltk.download('cess_esp') # para preeentenerar
from nltk.corpus import cess_esp as cess
from nltk import UnigramTagger as ut # etiquetador por Unigramas
from nltk import BigramTagger as bt # etiquetador por Bigramas
```

```python
#@title Entrenamiendo del tagger por unigramas
cess_sents = cess.tagged_sents() # para separar frases
fraction = int(len(cess_sents)*90/100) 
uni_tagger = ut(cess_sents[:fraction]) # definimos instancias de etiquetador pero solo le pasamos una fracion de el, aqui ya estamos entrenando el etiquetador
uni_tagger.evaluate(cess_sents[fraction+1:]) # evaluamos el resultado del modelo
# sobre el 10 % restante
# el resultado representa el % de precision del algoritmo para evaluar la frace
```

```python
uni_tagger.tag("Yo soy una persona muy amable".split(" ")) # ejemplo de etiquetar frase
```

```python
#@title Entrenamiento del tagger por bigramas, aunque el etiquetado pir unigramas es mejor
fraction = int(len(cess_sents)*90/100)
bi_tagger = bt(cess_sents[:fraction])
bi_tagger.evaluate(cess_sents[fraction+1:])
```

```python
bi_tagger.tag("Yo soy una persona muy amable".split(" "))
```

#EL modelo markoviano latente:  
El objetivo del modelo es encontrar dada una secuancia de palabas cual es la secuencia de etiquetas que le corresponden por mayor probabilidad

hipotesis de independencia:
$p(w^n/t^n)=\pi^n p(w_i / t_i)$

hipotesis Markoviana:
$p(t^n) = \pi^n *p(t_i /t_{i-1})=p(t_2/t1)p(t_3 /t_2)$
donde t son etiquetas y w son palabras
$t^n =argmax_{t^n}* \pi^n p(w_i /t_i)p(t_i/t_{i-1})$ con esto la secuencia de etiquetas mas probables esta dada por la multiplicacion de esas probabilidades que calcula HMM y ase todas las convinaciones posibles

## Entrenamiento del modelo - Calculo de conteos:

* tags (tags) `tagCountDict`: $C(tag)$
* emisiones (word|tag) `emissionProbDict`: $C(word|tag)$ dada una etiqueta cuantas veces le corresponde una palabra
* transiciones (tag|prevtag) `transitionDict`: $C(tag|prevtag)$

$P(tag|prevtag)=\dfrac{C(prevtag.tag)}{C(prevtag)}$

$P(word|tag)=\dfrac{C(word|tag)}{(tag)}$

```python
# bibliotecas
tagCountDict = {} # objeto de diccionario, cada elemento de diccionarioa es cuantas vecese aparece la etiqueta
emissionDict = {} # elemento de direciones, donde cade elemento es cuantas veces dada una etiqueta le corresponde una palabra
transitionDict = {} # elemento de direcciones, donde cada elemento es dada una etiqueta prebia cuantas veces le corresponde una etiquetea en la posision

tagtype = 'upos' # tagtype  para defirni el tipo de etiquetas que en este caso es upos
data_file = open("UD_Spanish-AnCora/es_ancora-ud-dev.conllu", "r", encoding="utf-8")

# Calculando conteos (pre-probabilidades)
for tokenlist in parse_incr(data_file):
  prevtag = None
  for token in tokenlist: # for para recorrido de la lista individual

    # C(tag) # empezamos a realizar conteos de todas las etiquetas
    tag = token[tagtype]
    if tag in tagCountDict.keys(): # definimos logica
      tagCountDict[tag] += 1
    else: # si la etiqueta no esta en el diccionario entonces creo la llave y le asigon un dccionario unico
      tagCountDict[tag] = 1

    # C(word|tag) -> probabilidades emision
    wordtag = token['form'].lower()+'|'+token[tagtype] # (word|tag) '|' parapegar
    if wordtag in emissionDict.keys():
      emissionDict[wordtag] = emissionDict[wordtag] + 1
    else:
      emissionDict[wordtag] = 1

    #  C(tag|tag_previo) -> probabilidades transición
    if prevtag is None:
      prevtag = tag
      continue # para saltar iteracion de ese for
    transitiontags = tag+'|'+prevtag # cojo la etiqueta actua y le pego la anterior
    if transitiontags in transitionDict.keys(): # si el elemento de etiqueta ya esta en el dic entonce al elemnto agregele una unidad mas  delocontrario crea la nueva llave y asignale el numero 1 
      transitionDict[transitiontags] = transitionDict[transitiontags] + 1
    else:
      transitionDict[transitiontags] = 1
    prevtag = tag # amedida que avanzo eltag anterior se vuelve el tag actual por eso prevtag = tag
    
transitionDict # para imprimir probabilidades de transicion
# emissionDict  # probabilidad de emision
# tagCountDict # conteo de diccionarios por etiquetas
```

# Entrenamiento del modelo - calculo de probabilidades

*   probabilidades de transición:
$$P(tag|prevtag) = \frac{C(prevtag, tag)}{C(prevtag)}$$

*   probabilidades de emisión:

 $$P(word|tag) = \frac{C(word|tag)}{C(tag)}$$


```python
transitionProbDict = {} # matriz A diccionario A y B
emissionProbDict = {} # matriz B

# transition Probabilities 
for key in transitionDict.keys():
  tag, prevtag = key.split('|')
  if tagCountDict[prevtag]>0:
    transitionProbDict[key] = transitionDict[key]/(tagCountDict[prevtag])
  else:
    print(key)

# emission Probabilities 
for key in emissionDict.keys():
  word, tag = key.split('|')
  if emissionDict[key]>0: # este if pregunta si en el dicionario de edicion la palabra 'key' tiene un conteo mayor que cero 
    emissionProbDict[key] = emissionDict[key]/tagCountDict[tag]
  else:
    print(key)

#esto nos muestra la probabilida de p(a/b) osea de que dado b se halle a
transitionProbDict['ADJ|ADJ'] # muestra probabilidades de transicion de ['ADJ|ADJ']  pero se puede solo hacer transitionProbDict y nos generara toda una lista
# emissionProbDict # muestra probabilidades de emision de que dado un tag la palabra sea x
```
